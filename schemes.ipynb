```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
g1=1
g2=1/2
l=1
```

```python
def real_solution(A: np.array):
  x=np.linspace(0,l,len(A))
  return g1-x+(g2-g1+l)*np.arctan(x)/np.arctan(l)
```

```python
A = np.fromfile("naive_linearization_test.txt", dtype = "float", sep = " ")
A = np.insert(A,0,g1)
A = np.append(A,g2)
xes = np.linspace(0,1, len(A))
```

```python
plt.plot(xes, A, label="Численное решение",linestyle = "-.")
plt.plot(xes, real_solution(A), label="Аналитическое решение", linestyle = "--")
plt.title("Решение уравнения теплопроводности")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
```

```python
plt.plot(xes, (A-real_solution(A))/real_solution(A))
plt.title("Ошибка численного решения уравнения теплопроводности")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```
